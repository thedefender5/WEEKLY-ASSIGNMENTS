# QUESTION 3 -  To check whether all the zones generate the same amount of sales.


ANOVA test is used as more than one independent variable is present to check whether all the zones generate the same amount of sales.

One way ANOVA test is used as these variables are compared with only the amount of sales.

```python
Taking the null hypothesis(H0) and alternate hypothesis(H1)
```

```python
H0 : All the zones generate the same amount of sales

H1 : All the zones generate different amount of sales
```

```python
from scipy.stats import f_oneway
```

# To  obtain the F statistic value and pvalue

```python
fval,pval=f_oneway(data1['Zone - A'],data1['Zone - B'],data1['Zone - C'],data1['Zone - D'])

print('F statistic is:',fval)
print('P value =',pval)
```

# CONCLUSION

```python
1. P value is less than 0.05,which is the significant value.

2. Therefore we reject the null hypothesis.

3. So we accept the alternate hypothesis.

4. Therefore it is clear that All the zones generate different amount of sales.
```
