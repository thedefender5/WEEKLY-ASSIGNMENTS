# 4 - To find the top 5 cars based on their sales price.


```python
print('Creating a data frame which involves only make, model and sell price of the cars')
```

```python
cars = pd.DataFrame({ 'Make': cars_data['Make'], 'Model':cars_data['Model'] , 'Sale Price':cars_data['Sale Price']})
cars
```

# Make and Model of the cars with highest sell price -top 5

```python
print('To display the list of top 5 cars with highest sell price')
```

```python
top_sales = cars.sort_values('Sale Price', ascending = False)
top_sales.head(5)
```
